```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import mnist
import numpy as np 
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(60000,784)
x_test = x_test.reshape(10000,784)
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
y_train[0].shape
```

```python
# model 을 생성하세요
# layer1 512개의 node , activation "relu"
# layer2 512개의 node , activation "relu"
# layer3 ??개의 node , activation "???????"

# loss = "categorial crossentropy", optimizer = adam , metric=["accuracy"]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(512, activation="relu", input_shape=(784,)))
model.add(tf.keras.layers.Dense(512, activation="relu"))
model.add(tf.keras.layers.Dense(10, activation="softmax"))
```

```python
model.compile(optimizer="adam", loss=tf.keras.losses.CategoricalCrossentropy(), metrics=["accuracy"])
```

```python
model.fit(x_train, y_train, batch_size=1024, epochs=10)
```

```python
model.predict(x_test[0])
```

```python
test_data = x_test[0].reshape(1,784)
```

```python
test_data.shape
```

```python
np.round(model.predict(test_data))
```
